## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1: Price Distribution
plt.figure(figsize=(10,5))
df['price'].plot(kind = 'hist')
plt.title('Distribution of Automobile Prices')
plt.xlabel('Price')
plt.ylabel('Count')

````

````python
sns.distplot(df['price'])
````

##### 1. Why did you pick the specific chart?

A histogram is the best starting point for understanding how the target variable is distributed.  
Since price is the main business variable, checking its spread helps identify whether the market is concentrated around affordable cars or premium vehicles.

##### 2. What is/are the insight(s) found from the chart?

The distribution is right-skewed, meaning most cars lie in the lower or mid-price range while a smaller number of high-end cars create a long upper tail.  
This shows that the dataset contains many economy or mid-range vehicles and fewer premium models.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. This insight helps in segmentation and pricing decisions.  
A company can create separate strategies for mass-market vehicles and premium vehicles instead of treating the full price range as one group.

#### Chart - 2

````python
# Chart - 2: Average Price by Make
avg_make = df.groupby('make')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
avg_make.plot(kind='bar')
plt.title('Average Price by Car Make')
plt.xlabel('Make')
plt.ylabel('Average Price')

````

##### 1. Why did you pick the specific chart?

A bar chart is suitable for comparing average price across brands.  
This helps identify which manufacturers are positioned as premium, mid-range, or economy players.


##### 2. What is/are the insight(s) found from the chart?

Brands like Jaguar, Mercedes-Benz, BMW, and Porsche show much higher average prices than brands such as Chevrolet, Dodge, or Plymouth.  
Brand identity clearly influences the market value of a vehicle.


##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Brand-based pricing insight is useful for competitors, dealers, and product managers.  
It supports premium positioning, competitor benchmarking, and targeted promotions.

#### Chart - 09 - Correlation Heatmap

````python
# Chart - 09 - Correlation Heatmap
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(14,10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
````

##### 1. Why did you pick the specific chart?

A heatmap is best for summarizing many numerical relationships at once.  
It helps find which features are strongly related to price and to each other.

##### 2. What is/are the insight(s) found from the chart?

Price is positively correlated with engine-size, horsepower, width, curb-weight, and wheel-base, while it is negatively correlated with city-mpg and highway-mpg.  
This means bigger, heavier, and more powerful cars cost more, while fuel-efficient cars tend to be cheaper.

#### Chart - 10 - Pair Plot

````python
# Chart - 10 - Pair Plot
pair_cols = ['price', 'horsepower', 'engine-size', 'curb-weight', 'city-mpg', 'highway-mpg']
sns.pairplot(df[pair_cols].dropna())
````

##### 1. Why did you pick the specific chart?

A pair plot gives a compact multi-variable view of distributions and pairwise relationships.  
It is useful for confirming patterns seen in individual charts.

##### 2. What is/are the insight(s) found from the chart?

The pair plot reinforces that price moves upward with horsepower, engine-size, and curb-weight, and downward with fuel efficiency measures.  
It also shows that several predictor variables are related to each other, which is important for future modeling.
