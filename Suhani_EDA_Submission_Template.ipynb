# **Project Summary -**

This Exploratory Data Analysis (EDA) project serves as a comprehensive investigation into the New York City Airbnb ecosystem using a dataset of approximately 48,895 listings from 2019. In an increasingly competitive short-term rental market, understanding the intersection of geography, pricing, and consumer behavior is vital for stakeholders. The primary objective of this study was to transform raw, unstructured data into a strategic narrative that assists hosts in optimizing their business models and enables guests to make data-driven booking decisions.

The analytical journey began with a rigorous Data Wrangling and cleaning phase. Recognizing that high-quality insights require clean data, I addressed missing values in the name and host_name columns to maintain dataset integrity. A critical step involved the reviews_per_month column; by filling null values with zero, I ensured that listings with no review history were correctly represented rather than being omitted. To prevent the analysis from being skewed by extreme values, I performed outlier detection, removing listings with a price of  0 and capping the upper limit at$1000. This created a more realistic "middle-market" view, which represents over 95% of the NYC Airbnb landscape.

Through Univariate Analysis, several foundational truths emerged. The NYC market is essentially a duopoly of room types: Entire homes/apartments (52%) and Private rooms (46%). Shared rooms are statistically insignificant, representing less than 2% of total supply. When examining pricing, the distribution showed a heavy right skew, with the "sweet spot" for listings falling between 50 and $150. Interestingly, the analysis of host behavior through the availability_365 variable revealed a bimodal distribution. This signifies a market split between "casual hosts" who rent their homes for less than 50 days a year and "professional operators" who maintain near 100% availability.

The Bivariate and Multivariate phases provided deeper strategic layers. Geographically, Manhattan remains the undisputed premium hub with an average price of ~197, while the Bronx offers the most affordable entry point at ~87. A key discovery from the scatter plots and boxplots was the inverse relationship between price and review volume; listings priced under $150 accumulate significantly more reviews, suggesting that affordability is the primary driver of high turnover and platform visibility. Furthermore, while Manhattan leads in total revenue potential, Brooklyn shows a more diverse mix of room types and a high density of listings in neighborhoods like Williamsburg and Bedford-Stuyvesant, making it the most competitive sub-market for value-conscious travelers.

Finally, the correlation analysis demonstrated that while price is influenced by location and room type, it has a weak correlation with availability_365, suggesting that simply lowering prices does not guarantee a listing will be available more often—it must be paired with high-demand locations. The insights gained from these 15+ visualizations provide a roadmap for market entry: new hosts should target the $ 50– $150 range with a 1–3 night minimum stay to maximize their "reviews per month" and climb the platform rankings. Overall, the project highlights that while Manhattan offers the highest "ceiling" for luxury revenue, the true engine of the NYC Airbnb market lies in the affordable, high-turnover listings found within the vibrant neighborhoods of Brooklyn and Queens.

### Dataset Loading

````python

1. IMPORTING LIBRARIES

import pandas as pd
import os

2. LOADING DATA
def load_airbnb_data(file_name):
    try:
        if os.path.exists(file_name):
            data = pd.read_csv(file_name)
            print("Success: Dataset loaded successfully!")
            return data
        else:
            print(f"Error: The file '{file_name}' was not found in the directory.")
            return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# Usage
file_path = 'Copy of Copy of Airbnb NYC 2019.csv'
df = load_airbnb_data(file_path)
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Remove duplicates
df.drop_duplicates(inplace=True)

# Drop unnecessary columns
df.drop(['id', 'host_name', 'last_review'], axis=1, inplace=True)

# Fill missing values
df['reviews_per_month'].fillna(0, inplace=True)
df['name'].fillna('No Name', inplace=True)

# Remove price outliers
df = df[df['price'] > 0]
df = df[df['price'] < 1000]

print("Final shape:", df.shape)
df.head()
````

### What all manipulations have you done and insights you found?

1. Removed duplicate rows
2. Dropped id, host_name, last_review columns (not needed for analysis)
3. Filled missing reviews_per_month with 0
4. Filled missing name with 'No Name'
5. Removed listings with price = 0 or price > 1000 (outliers)
Final dataset is clean and ready for visualization.
